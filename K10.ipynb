```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('k10.csv')
```

```python
df
```

```python
df = df.sort_values(by=['Company', 'Year'])
```

```python
# Function to calculate Debt to Assets (%)
def calculate_debt_to_assets(df):
    df['Debt to Assets (%)'] = (df['Total Liabilities'] / df['Total Assets']) * 100
    return df

def calculate_cash_flow_to_revenue(df):
    df['Cash Flow to Revenue (%)'] = (df['Cash Flow'] / df['Total Revenue']) * 100
    return df

def calculate_yoy_growth(df, column_name):
    df[f'{column_name} Growth (%)'] = df.groupby('Company')[column_name].pct_change() * 100
    return df

# Apply the functions
df = calculate_debt_to_assets(df)
df = calculate_cash_flow_to_revenue(df)
df = calculate_yoy_growth(df, 'Total Revenue')
df = calculate_yoy_growth(df, 'Net Income')
df = calculate_yoy_growth(df, 'Total Assets')
df = calculate_yoy_growth(df, 'Total Liabilities')
df = calculate_yoy_growth(df, 'Cash Flow')
```

```python
df.fillna(0, inplace=True)
```

```python
df
```

```python
df.to_csv('k10_trends.csv')
```

```python
# 1. Sum of financial metrics by Company
sum_by_company = df.groupby('Company').sum()

# 2. Average of financial metrics by Company
mean_by_company = df.groupby('Company').mean()

# 3. Median of financial metrics by Company
median_by_company = df.groupby('Company').median()

# 4. Max values by Year
max_by_year = df.groupby('Year').max()

# 5. Min values by Year
min_by_year = df.groupby('Year').min()

# 6. Standard Deviation by Year
std_by_year = df.groupby('Year').std(numeric_only=True)

# 7. Cumulative sum of Total Revenue by Year
df['Cumulative Revenue'] = df.groupby('Company')['Total Revenue'].cumsum()

# 8. Rolling average of Total Revenue by Year (window size = 2)
df['Rolling Revenue'] = df.groupby('Company')['Total Revenue'].rolling(window=2).mean().reset_index(level=0, drop=True)

# 9. Pivot table: Average Total Revenue by Company and Year
pivot_table = pd.pivot_table(df, values='Total Revenue', index='Company', columns='Year', aggfunc='mean')
```

```python

# Set the style for all plots
sns.set_style("whitegrid")
plt.style.use("seaborn-v0_8")


# Create a figure with multiple subplots
fig = plt.figure(figsize=(5, 5))

# 1. Revenue Trends
plt.plot()
sns.lineplot(data=df, x='Year', y='Total Revenue', hue='Company', marker='o', linewidth=2.5)
plt.title('Revenue Trends (2022-2024)', fontsize=12, pad=15)
plt.ylabel('Revenue (Millions USD)')
plt.show()
```

Shows increasing revenue trends for all companies, with Apple consistently leading.

```python
# 2. Net Income Comparison
plt.plot()
sns.barplot(data=df, x='Year', y='Net Income', hue='Company')
plt.title('Net Income Comparison', fontsize=12, pad=15)
plt.ylabel('Net Income (Millions USD)')
plt.show()
```

Highlights profitability, with Apple maintaining the highest net income, and Tesla showing significant growth.

```python

# 3. Assets vs Liabilities
plt.plot()
metrics = ['Total Assets', 'Total Liabilities']
for company in df['Company'].unique():
    company_data = df[df['Company'] == company]
    plt.plot(company_data['Year'], company_data['Total Assets'], marker='o', label=f'{company} Assets')
    plt.plot(company_data['Year'], company_data['Total Liabilities'], marker='s', linestyle='--', 
             label=f'{company} Liabilities')
plt.title('Assets vs Liabilities Trends', fontsize=12, pad=15)
plt.ylabel('Amount (Millions USD)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Illustrates the financial stability of each company by comparing their total assets and liabilities.

```python
# 4. Cash Flow Analysis
plt.plot()
sns.lineplot(data=df, x='Year', y='Cash Flow', hue='Company', marker='o')
plt.title('Cash Flow Trends', fontsize=12, pad=15)
plt.ylabel('Cash Flow (Millions USD)')
plt.show()
```

Depicts operational liquidity, where Microsoft and Apple display strong and steady cash flows.

```python
# 5. Profitability Ratios
plt.plot()
df['Profit Margin'] = (df['Net Income'] / df['Total Revenue']) * 100
sns.barplot(data=df, x='Company', y='Profit Margin', hue='Year')
plt.title('Profit Margins by Company', fontsize=12, pad=15)
plt.ylabel('Profit Margin (%)')
plt.show()
```

Reveals Apple's superior profitability, with Tesla showing an improving margin over the years.

```python
# 6. Asset Efficiency
plt.plot()
df['Asset Turnover'] = df['Total Revenue'] / df['Total Assets']
sns.barplot(data=df, x='Company', y='Asset Turnover', hue='Year')
plt.title('Asset Turnover Ratio', fontsize=12, pad=15)
plt.ylabel('Asset Turnover (times)')
plt.show()
```

Indicates efficiency in using assets to generate revenue.

```python
# 7. Cash Flow to Revenue Ratio
plt.plot()
df['Cash Flow Ratio'] = (df['Cash Flow'] / df['Total Revenue']) * 100
sns.barplot(data=df, x='Company', y='Cash Flow Ratio', hue='Year')
plt.title('Cash Flow to Revenue Ratio', fontsize=12, pad=15)
plt.ylabel('Cash Flow Ratio (%)')

plt.show()
```

The Cash Flow to Revenue Ratio is a measure of how effectively a company generates cash from its revenues. Microsoft as the most financially efficient company, generating substantial cash relative to its revenue while Tesla's lower ratio highlights its focus on aggressive growth strategies

```python
# 8. Financial Metrics Heatmap
plt.plot()
pivot_data = df.pivot_table(
    values=['Total Revenue', 'Net Income', 'Cash Flow'],
    index='Company',
    columns='Year'
)
sns.heatmap(pivot_data, annot=True, fmt='.0f', cmap='YlOrRd')
plt.title('Financial Metrics Heatmap', fontsize=12, pad=15)

plt.show()
```

```python

# Create radar chart
def create_radar_chart(companies, metrics, values, year):
    num_metrics = len(metrics)
    angles = [n / float(num_metrics) * 2 * np.pi for n in range(num_metrics)]
    angles += angles[:1]
    
    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(projection='polar'))
    
    for i, company in enumerate(companies):
        company_values = values[i]
        company_values += company_values[:1]
        ax.plot(angles, company_values, linewidth=1, linestyle='solid', label=company)
        ax.fill(angles, company_values, alpha=0.25)
    
    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    ax.set_rlabel_position(0)
    
    plt.xticks(angles[:-1], metrics)
    ax.set_title(f'Company Performance Comparison ({year})', pad=20)
    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
    
    return fig

# Prepare data for 2024
companies = df[df['Year'] == 2024]['Company'].unique()
metrics = ['Revenue', 'Net Income', 'Assets', 'Cash Flow', 'Profit Margin']

# Normalize values for radar chart
latest_data = df[df['Year'] == 2024]
values = []
for company in companies:
    company_data = latest_data[latest_data['Company'] == company]
    normalized_values = [
        company_data['Total Revenue'].values[0] / latest_data['Total Revenue'].max(),
        company_data['Net Income'].values[0] / latest_data['Net Income'].max(),
        company_data['Total Assets'].values[0] / latest_data['Total Assets'].max(),
        company_data['Cash Flow'].values[0] / latest_data['Cash Flow'].max(),
        company_data['Profit Margin'].values[0] / latest_data['Profit Margin'].max()
    ]
    values.append(normalized_values)

# Create radar chart for 2024
radar_fig = create_radar_chart(companies, metrics, values, 2024)
plt.show()
```
